```python
df = pd.DataFrame()
for i in range(1,7):
# for i in [3,4,5]:
    fname = 'iit'+str(i)+'.pkl'
    df_i = pd.read_pickle(fname)
    df_i = df_i.drop(columns=['sub', 'cg'])
    if len(df)==0:
        df = df_i
    else:
        df = pd.concat([df, df_i], ignore_index=True)

df = df[df['single_amount']%100==0]
df['rate_adjust_pc_pt'] = 100*df['rate_adjust']
```
